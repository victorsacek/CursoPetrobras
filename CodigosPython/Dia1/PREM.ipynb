```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r,rho = np.loadtxt("https://raw.githubusercontent.com/victorsacek/CursoPetrobras/master/densidade_radial.txt"
          ,unpack=True)
```

```python
plt.plot(r,rho,".-")
plt.xlabel("Distância radial (km)")
plt.ylabel("Densidade (kg/m$^3$)")
```

```python
r_interp = np.linspace(0.0,6371.0,600)
rho_interp = np.interp(r_interp,r,rho)

G = 6.67E-11 # Constante Gravitacional Universal
r_interp = r_interp*1000 # Converto km para m

dr = r_interp[1]-r_interp[0]

g = 0.0 # Aceleração gravitacional no centro da Terra
M = 0.0 # Massa inicial

n = len(r_interp)

for i in range(1,n):
  M = M + rho_interp[i]*4*np.pi*r_interp[i]**2*dr

  g = np.append(g,G*M/r_interp[i]**2)

#print(g)

plt.plot(r_interp/1000,g)
plt.grid()

plt.ylabel("Aceleração da gravidade (m/s$^2$)")
plt.xlabel("Raio (km)")

r_nucleo = 6371-2900.0
plt.plot([r_nucleo,r_nucleo],[0,11],"k--")
```
