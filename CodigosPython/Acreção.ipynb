```python
import numpy as np
```

```python
M = 6.0E24 # Massa da Terra 6 x 10^24 kg
G = 6.67E-11 # (m3 kg-1 s-2) Constante Gravitacional Universal (SI)
R = 6_371_000.0 # Raio médio da Terra (m)

ve = np.sqrt(2*G*M/R)

print(ve/1000,"km/s")
```

```python
rho = 3400.0 # (kg/m3) Densidade do planetesimal
d = 10_000.0 # Diâmetro do corpo

r = d/2

V = (4./3.)*np.pi*r**3 # Volume do corpo

m = rho*V

c = 750.0 # J/kg/K calor específico para a Terra

DeltaT = m*ve**2/(2*M*c)

print("%.3g oC"%(DeltaT))
print("%.6f oC"%(DeltaT))
```

```python
n = M/m
print(n)
print("%.2f oC"%(DeltaT*n))
```
