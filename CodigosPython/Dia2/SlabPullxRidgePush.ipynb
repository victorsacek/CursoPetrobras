```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rho0 = 3300.0 # Densidade de referência para o manto (kg/m3)
alfa = 3.28E-5 # Coeficiente de expansão volumétrico (1/oC)

Tasteno = 1300.0 # Temperatura do manto astenosférico (oC)
Tslab = 750.0 # Temperatura média do slab (oC)

drho = rho0*alfa*(Tasteno-Tslab)
print(drho,"kg/m3 - contraste de densidade entre a placa e o manto circundante")

g = 10.0 # Gravidade (m/s2)
L = 700_000.0 # Comprimento da placa (m)
b = 100_000.0 # Espessura da placa (m)

Fsp = drho*L*b*g

print("%.2g N/m"%(Fsp))
```

```python
# Força relacionada com o Ridge Push

dz = 200.0 # espaçamento entre os pontos
zz = np.arange(0.0,b,dz)

P1 = rho0*g*zz # Pressão em diferentes profundidades sob a dorsal
F1 = np.sum(P1*dz)

rhow = 1000.0 # Densidade da água
w = 2500.0 # Diferenca da espessura da lâmina dagua entre dorsal e fundo do mar

rho2 = zz*0 + rho0
# Criamos um vetor em que todos os elementos possuem o mesmo valor rho0
rho2[zz<w] = rhow

P2 = np.cumsum(rho2*g*dz)

F2 = np.sum(P2)*dz

print("Magnitude do ridge push = %.2g N/m"%(F1-F2))

```
