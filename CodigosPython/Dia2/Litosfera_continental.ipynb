```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100_000.0,1000.0)

H = 9.6E-10
# (W/kg) Produção de calor radiogênico por unidade de massa

rho = 2800.0 # Densidade (kg/m3)

q0 = 0.05 # (W/m2) Fluxo térmico superficial
k = 2.0 # Condutividade térmica das rochas (W/(m.K))
T0 = 20.0 # Temperatura superficial (oC)

T = T0 + q0*x/k - rho*H*x**2/(2*k)

plt.plot(T,x/1000)
plt.ylim(100,0)
plt.xlabel(r"Temperatura ($^\circ$C)")
plt.ylabel("Profundidade (km)")
```

```python
qm = 0.025 #W/m2 (fluxo de calor do manto)
hr = 10_000.0
# m (Fator de escala da diminuição da produção de calor radio. com a profundidade)
rhoH0 = 2.0E-6 # W/m3 (Produção de calor por volume)

T = T0 + qm*x/k + hr**2*rhoH0*(1.0-np.exp(-x/hr))/k

plt.plot(T,x/1000)
plt.ylim(100,0)
plt.xlabel(r"Temperatura ($^\circ$C)")
plt.ylabel("Profundidade (km)")
```
