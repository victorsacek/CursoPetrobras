```python
rho = 4000.0 # densidade adotada (kg/m3)
g = 10.0 # aceleração gravitacional (m/s2)
alfa = 3.0E-5 # Coeficiente de expansão volumétrico (1/K)
DeltaT = 1700.0 # Constraste de temperatura (K) entre a superfície e a base do manto superior
b = 700_000.0 # Espessura do manto superior (m)
visc = 1.0E21 # Pa.s viscosidade
kappa = 1.0E-6 # Difusividade térmica

Ra = rho*g*alfa*DeltaT*b**3/(visc*kappa)

print(Ra)
```
